### 2.1. Группы по срокам финансирования

Разделите все компании на три группы:

* Единичное финансирование — был всего один раунд финансирования.

* Срок финансирования до года — между первым и последним раундом финансирования прошло не более года.

* Срок финансирования более года.

Визуализируйте соотношение этих групп, создав два графика:

* По количеству компаний: Покажите, какой процент от общего числа компаний относится к каждой из трёх групп.
* По объёму инвестиций: Отобразите, какую долю от общего объёма привлечённых средств получила каждая группа.

Совет: Для ясности и согласованности используйте единую цветовую палитру для всех графиков, чтобы каждая категория (например, «Единичное финансирование») всегда отображалась одним цветом.

````python
# Создаем копию датасета для работы
df = investments.copy()

# Определяем группы по срокам финансирования
def funding_group(row):
    # Единичное финансирование
    if row['funding_rounds'] == 1:
        return 'Единичное финансирование'
    
    # Если есть даты первого и последнего финансирования
    if pd.notna(row['first_funding_at']) and pd.notna(row['last_funding_at']):
        days_diff = (row['last_funding_at'] - row['first_funding_at']).days
        
        # Срок до года (включительно)
        if days_diff <= 365:
            return 'Срок до года'
        # Срок более года
        else:
            return 'Срок более года'

# Применяем функцию
df['funding_period'] = df.apply(funding_group, axis=1)

# Проверяем распределение
print("Распределение компаний по группам:")
print(df['funding_period'].value_counts())
print("\nВ процентах:")
print(df['funding_period'].value_counts(normalize=True).mul(100).round(2))
````

Определите компании с аномальным объёмом общего финансирования — используйте метод IQR отдельно по каждому сегменту. Напомним, что все нишевые сегменты должны быть объединены в одну группу, а средние — в другую.

Определите сегменты рынка с наибольшей долей компаний, получивших аномальное финансирование, и выведите топ таких сегментов.

**Анализ по сегментам необходим, чтобы учитывать разный масштаб инвестиций в разных отраслях и корректно определять аномалии относительно конкретного рынка, а не всех компаний подряд.**

````python
# 1. Создаем столбец для выбросов
df['is_outlier'] = False
# 2. Для каждого сегмента считаем верхнюю границу IQR и отмечаем выбросы
for segment in df['market'].unique():
    data = df[df['market'] == segment]['funding_total_usd']
    if len(data) > 3:  # проверяем, что данных достаточно
        Q3 = data.quantile(0.75)
        IQR = Q3 - data.quantile(0.25)
        upper = Q3 + 1.5 * IQR
        # Отмечаем компании с суммой выше границы
        df.loc[(df['market'] == segment) & (df['funding_total_usd'] > upper), 'is_outlier'] = True
# 3. Считаем долю выбросов по сегментам
stats = []
for segment in df['market'].unique():
    data = df[df['market'] == segment]
    stats.append({
        'сегмент': segment,
        'выбросов': data['is_outlier'].sum(),
        'всего': len(data),
        'доля': round(data['is_outlier'].mean(), 3)
    })
# 4. Сортируем и показываем топ-15
result = pd.DataFrame(stats).sort_values('доля', ascending=False).reset_index(drop=True)
print("ТОП-15 СЕГМЕНТОВ ПО ДОЛЕ ВЫБРОСОВ:")
display(result.head(15))
# 5. ОБЩАЯ СТАТИСТИКА ПО ВСЕМ ВЫБРОСАМ
total_outliers = df['is_outlier'].sum()
total_companies = len(df)
outlier_percent = (total_outliers / total_companies) * 100

print("\n" + "="*60)
print("ОБЩАЯ СТАТИСТИКА ПО ВЫБРОСАМ")
print("="*60)
print(f"Всего компаний в датасете: {total_companies}")
print(f"Всего выявлено компаний-выбросов: {total_outliers}")
print(f"Доля выбросов: {outlier_percent:.2f}%")
# 6. График для топ-15
plt.figure(figsize=(14, 8))
top = result.head(15)['сегмент'].tolist()
plot_data = df[df['market'].isin(top)].copy()
# Сортируем сегменты на графике
plot_data['market'] = pd.Categorical(plot_data['market'], categories=top[::-1], ordered=True)
sns.boxplot(data=plot_data, y='market', x='funding_total_usd')
plt.xscale('log')
plt.xlabel('Объем финансирования (log scale)')  # подпись оси X
plt.ylabel('Сегмент рынка')                       # подпись оси Y
plt.title('Топ-15 сегментов с наибольшей долей выбросов')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
````

__Выводы по анализу выбросов в сегментах__

__Общая статистика:__

При этом слишком большие аномальные значения, то есть неадекватные выбросы, нужно заменить на пропуски.

Совет: когда будете делить сумму возвращённых средств на суммарный объём привлечённого финансирования по конкретному году, добавьте к знаменателю небольшое число, например `1e-60`. Это поможет избежать деления на ноль.

````python
# 1. Подготавливаем данные по инвестициям (суммарно по годам)
# Группируем инвестиции по годам для каждого типа финансирования
investment_by_year = df_final.groupby('year')[['venture', 'debt_financing', 'private_equity', 'seed', 'angel']].sum()

print("ИНВЕСТИЦИИ ПО ГОДАМ (млн $):")
display(investment_by_year / 1e6)

# 2. Данные по возвратам (уже в долларах)
print("\nВОЗВРАТЫ ПО ГОДАМ (млн $):")
display(returns / 1e6)

# 3. Создаем DataFrame для долей возврата
# Находим общие года в обоих датасетах
common_years = sorted(set(investment_by_year.index) & set(returns.index))
print(f"\nОбщие года для анализа: {common_years}")

# Создаем таблицу для долей возврата
return_ratios = pd.DataFrame(index=common_years)
funding_types = ['venture', 'debt_financing', 'private_equity', 'seed', 'angel']

for funding_type in funding_types:
    ratios = []
    for year in common_years:
        # Инвестиции в этом году
        inv = investment_by_year.loc[year, funding_type]
        # Возвраты в этом году (из датасета returns)
        ret = returns.loc[year, funding_type]
        
        # Добавляем 1e-60 к знаменателю для избежания деления на 0
        ratio = ret / (inv + 1e-60)
        
        # Заменяем аномальные значения (>1 или <0) на NaN 
        #(ratio > 1 Нельзя вернуть больше, чем было инвестировано ( >100%);
        # ratio < 0	Возврат не может быть отрицательным
        if ratio > 1 or ratio < 0:
            ratios.append(np.nan)
        else:
            ratios.append(round(ratio * 100, 2))  # в процентах
    
    return_ratios[funding_type] = ratios

print("\nДОЛЯ ВОЗВРАТОВ ПО ГОДАМ (%):")
display(return_ratios)
````

__Выводы по динамике доли возвратов__ 

- Venture — наиболее стабильный тип со средней долей возврата 30-40% и ростом до 78.6% в 2014.

- Debt_financing — высокая волатильность (22-92%), пик в 2008 (92.8%).

- Private_equity — стабильно высокие показатели (50-90%), но данные обрываются в 2014.

- Seed — устойчивое снижение с 99.7% (2000) до 14.8% (2013).

- Angel — нестабилен, много пропусков, лучший результат 83.4% (2004).
