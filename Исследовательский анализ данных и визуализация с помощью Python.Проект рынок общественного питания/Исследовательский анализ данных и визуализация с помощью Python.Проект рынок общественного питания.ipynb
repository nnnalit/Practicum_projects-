	# Оптимизация числовых типов данных
    def optimize_memory_usage(df: pd.DataFrame, print_size: bool=True) -> pd.DataFrame:
        '''
        Function optimizes memory usage in dataframe
        df: pd.DataFrame - data table
        print_size: bool - display of optimization results
        return pd.DataFrame - amount of optimized memory
        '''
        numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64'] # Типы, которые будем проверять на оптимизацию
        # Размер занимаемой памяти до оптимизации (в Мб)
        before_size = df.memory_usage().sum() / 1024**2    
        for column in df.columns:
            column_type = df[column].dtypes
            if column_type in numerics:
                column_min = df[column].min()
                column_max = df[column].max()
                if str(column_type).startswith('int'):
                    if column_min > np.iinfo(np.int8).min and column_max < np.iinfo(np.int8).max:
                        df[column] = df[column].astype(np.int8)
                    elif column_min > np.iinfo(np.int16).min and column_max < np.iinfo(np.int16).max:
                        df[column] = df[column].astype(np.int16)
                    elif column_min > np.iinfo(np.int32).min and column_max < np.iinfo(np.int32).max:
                        df[column] = df[column].astype(np.int32)
                    elif column_min > np.iinfo(np.int64).min and column_max < np.iinfo(np.int64).max:
                        df[column] = df[column].astype(np.int64)  
                else:
                    if column_min > np.finfo(np.float32).min and column_max < np.finfo(np.float32).max:
                        df[column] = df[column].astype(np.float32)
                    else:
                        df[column] = df[column].astype(np.float64)    
        # Размер занимаемой памяти после оптимизации (в Мб)
        after_size = df.memory_usage().sum() / 1024**2
        if print_size: print('Размер использования памяти: до {:5.2f} Mb - после {:5.2f} Mb ({:.1f}%)'
                             .format(before_size, after_size, 100 * (before_size - after_size) / before_size))
        return df
                                                                                                   
                                                                                                   
Проверяет минимальное и максимальное значения в столбце. Если значения могут уместиться в типе данных меньшего размера (например, np.int8 вместо np.int16), столбец конвертируется в более компактный тип данных. И для оценки стоило всё это начинать, подсчитывается размер занимаемой памяти DataFrame до и после оптимизации и выводит разницу в процентах

````python
# Определяем функцию, которая определит, работает ли заведение круглосуточно
def is_24_7(hours_series):
    """
    Функция определяет, работает ли заведение круглосуточно (24/7).
    Возвращает True, если заведение работает ежедневно и круглосуточно, и False в противоположном случае.
    """
    # Приводим к нижнему регистру
    hours = hours_series.str.lower().fillna('')
    
    # Круглосуточность
    round_the_clock = hours.str.contains('круглосуточно', na=False)
    
    # Ежедневность (два варианта)
    daily_1 = hours.str.contains('ежедневно', na=False)
    daily_2 = hours.str.contains('пн-вс', na=False)
    
     # Вариант 3: все дни недели перечислены
    days = ['пн', 'вт', 'ср', 'чт', 'пт', 'сб', 'вс']
    all_days = pd.Series(True, index=hours.index)
    for day in days:
        all_days = all_days & hours.str.contains(day, na=False)
    
    # Объединяем все варианты ежедневности
    is_daily = daily_1 | daily_2 | all_days
    
    # Финальный признак: И круглосуточно, И ежедневно
    return round_the_clock & is_daily

# Применяем
df['is_24_7'] = is_24_7(df['hours'])
````

Исследуем соотношение сетевых и несетевых заведений в разрезе категорий.

````python
# Создаём нормализованную таблицу с долями (процентами) по каждой категории
# Используем normalize='index' для получения процентов по строкам
chain_norm = pd.crosstab(df['category'], df['chain'], normalize='index') * 100
chain_norm.columns = ['Несетевые (%)', 'Сетевые (%)']

# Сортируем по доле сетевых заведений (чтобы категории с большей долей были сверху)
chain_norm = chain_norm.sort_values('Сетевые (%)', ascending=False)
display(chain_norm)
````

Так как вопрос сформулирован как "какие категории чаще являются сетевыми", мы сортируем данные по процентной доле сетевых заведений в каждой категории, а не по их общему количеству. Абсолютные цифры не показывают частоту встречаемости — в крупных категориях сетевых заведений может быть много даже при низкой доле. 

<div class="alert alert-info"> <b>Исправлено V8:</b> Обновляю задачу 3.3 "Сетевые vs несетевые заведения" с использованием нормализованных данных, стековой диаграммы и подписями процентов. </div>

````python
# Строим стековую горизонтальную диаграмму
plt.figure(figsize=(12, 8))

# Для горизонтальной диаграммы удобнее сортировать по возрастанию
# (чтобы категории с большей долей были сверху)
plot_data = chain_norm.sort_values('Сетевые (%)', ascending=True)

# Строим стековую диаграмму
ax = plot_data.plot(kind='barh', stacked=True,
                    color=['#F08080', '#48D1CC'],
                    figsize=(12, 8),
                    width=0.8,
                    legend=False,
                    ax=plt.gca())

# Добавляем подписи с процентами на столбцы
for i, (category, values) in enumerate(plot_data.iterrows()):
    non_chain = values['Несетевые (%)']
    chain = values['Сетевые (%)']
    
    # Подпись для несетевых (слева)
    if non_chain > 5:  # Подписываем только если доля больше 5%
        plt.text(non_chain/2, i, f"{non_chain:.1f}%", 
                ha='center', va='center', fontweight='bold', color='black', fontsize=10)
    
    # Подпись для сетевых (справа)
    if chain > 5:  # Подписываем только если доля больше 5%
        plt.text(non_chain + chain/2, i, f"{chain:.1f}%", 
                ha='center', va='center', fontweight='bold', color='black', fontsize=10)

# Добавляем легенду справа от графика
plt.legend(['Несетевые', 'Сетевые'], bbox_to_anchor=(1.15, 1), loc='upper left', fontsize=11)

# Настраиваем оформление
plt.title('Соотношение сетевых и несетевых заведений по категориям', fontsize=14, pad=15)
plt.xlabel('Доля (%)', fontsize=12)
plt.ylabel('Категория заведения', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
````

В целом по рынку `несетевых` заведений значительно больше, чем `сетевых`: `4779 заведения (61%) против 3088 (39%)`. Это говорит о том, что рынок общественного питания Москвы представлен в основном независимыми заведениями.

Изучите, с какими данными показывают самую сильную корреляцию рейтинги заведений? Чтобы понять, существует ли связь между рейтингом заведений и другими характеристиками, используем матрицу корреляции. В Python существует универсальная библиотека phik, которая помогает изучать связи между данными разного типа — как непрерывными, так и категориальными.

Библиотеку phik можно использовать для анализа связи между рейтингом заведения и другими признаками. В наших данных представлены разные по природе значения: непрерывные (количество мест, рейтинг), бинарные (сетевое/несетевое, круглосуточная работа) и категориальные (категория заведения, район, ценовая категория).

Рассчитаем матрицу корреляции phi_k для данных о заведениях общественного питания Москвы. Для составления матрицы корреляции отберём данные, которые потенциально могут влиять на рейтинг заведения: категория заведения, административный район, статус сетевого заведения, количество посадочных мест, ценовая категория и признак круглосуточной работы.

````python
# Вычисляем корреляционную матрицу с использованием phi_k
correlation_matrix = df[['category', 'district', 'chain', 'seats', 
                         'price', 'is_24_7', 'rating']].phik_matrix()

# Выводим результат
print('Корреляционная матрица с коэффициентом phi_k для переменной rating')
correlation_matrix.loc[correlation_matrix.index != 'rating'][['rating']].sort_values(by='rating', ascending=False)
````

Нас интересует связь рейтинга заведения с другими данными, поэтому на экран выведем только столбец rating итоговой матрицы корреляции. Условие correlation_matrix.loc[correlation_matrix.index != 'rating'] помогает убрать строку с индексом 'rating'.

Корреляционный анализ с использованием коэффициента phi_k показал, что рейтинг заведений имеет слабые связи с большинством исследованных факторов. Все коэффициенты корреляции находятся в диапазоне от 0.00 до 0.22, что говорит об отсутствии сильных зависимостей.

Самый сильный фактор из имеющихся — ценовая категория, но и его влияние слабое (0.223). Это может говорить о том, что посетители ожидают более высокого качества от дорогих заведений и в целом получают его, но зависимость далеко не линейная.

Корреляционный анализ показал, что самый сильный фактор, влияющий на рейтинг, — ценовая категория (коэффициент phi_k = 0.221). Однако phi_k показывает только наличие связи, но не её направление. Проверим эту связь детальнее.

````python
# Проверяем связь между ценовой категорией и рейтингом
price_rating = df.dropna(subset=['price', 'rating']).groupby('price')['rating'].agg(['mean', 'median', 'count', 'std']).round(2)
price_rating = price_rating.sort_values('mean', ascending=False)

print("Средний рейтинг по ценовым категориям:")
display(price_rating)
````

````python
# Визуализация зависимости рейтинга от ценовой категории
plt.figure(figsize=(12, 5))

# График 1: средние значения с доверительными интервалами
plt.subplot(1, 2, 1)
order = ['низкие', 'средние', 'выше среднего', 'высокие']
sns.barplot(data=df.dropna(subset=['price', 'rating']), 
            x='price', y='rating', 
            order=order,
            palette='viridis',
            ci=95)  # 95% доверительный интервал

plt.title('Средний рейтинг по ценовым категориям\n(с 95% доверительным интервалом)', fontsize=12)
plt.xlabel('Ценовая категория')
plt.ylabel('Средний рейтинг')
plt.grid(axis='y', alpha=0.3)

# График 2: распределение рейтингов (boxplot)
plt.subplot(1, 2, 2)
sns.boxplot(data=df.dropna(subset=['price', 'rating']), 
            x='price', y='rating', 
            order=order,
            palette='viridis')

plt.title('Распределение рейтингов по ценовым категориям', fontsize=12)
plt.xlabel('Ценовая категория')
plt.ylabel('Рейтинг')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# Кодируем ценовые категории в порядковые значения для корреляции Кендалла
price_mapping = {'низкие': 1, 'средние': 2, 'выше среднего': 3, 'высокие': 4}

df_temp = df.dropna(subset=['price', 'rating']).copy()
df_temp['price_code'] = df_temp['price'].map(price_mapping)

from scipy.stats import kendalltau

corr, p_value = kendalltau(df_temp['price_code'], df_temp['rating'])

print(f"Корреляция Кендалла между ценой и рейтингом: {corr:.3f}")
print(f"p-value: {p_value:.4f}")
print(f"Статистическая значимость: {'Да' if p_value < 0.05 else 'Нет'}")
````

Корреляция Кендалла 0.229 указывает на умеренную положительную связь между ценой и рейтингом. Посетители ожидают более высокого качества от дорогих заведений и в целом получают его, о чём свидетельствуют более высокие оценки.
